```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
orders = pd.read_csv('orders-csv.csv')
products = pd.read_csv('products.csv')
order_products_prior = pd.read_csv('order_products__prior.csv')
aisles = pd.read_csv('aisles.csv')
orders_hour = orders.groupby('order_hour_of_day')['order_number'].count().to_frame('number_of_orders_each_hour')
plt.figure(figsize=(10,7))
sns.barplot(orders_hour.index, orders_hour.number_of_orders_each_hour, order=orders_hour.index)
plt.xlabel('hour of the day', size=15)
plt.ylabel('Total orders per hour', size=15)
plt.xticks(size=12, rotation=90)
plt.show()
```

```python
total_orders= orders_hour.number_of_orders_each_hour.sum()
orders9_5=orders_hour.number_of_orders_each_hour[9:18].sum()
pct = (orders9_5/total_orders)*100
print('The orders placed from 9 to 5 are the '+ str(pct) +'% from total orders')
```

```python
prior = orders[orders.eval_set == 'prior']
prior = pd.merge(prior, order_products_prior, on = 'order_id' ,how='left' )
prior.head()
```

```python
product_avg = prior.groupby('product_id')['add_to_cart_order'].mean().to_frame('avg_add_to_cart_pos')
product_avg = product_avg.reset_index()
product_avg.head()
```

```python
orders_prior = orders[orders.eval_set == 'prior']
orders_prior.head()
```

```python
orders_25 = orders_prior.groupby('user_id').filter(lambda x: x.order_number.max() >= 25)
orders_25.head()
```

```python
orders_25_avg = orders_25.groupby('user_id')['days_since_prior_order'].mean().to_frame('avg_days_order_since_25')
orders_25_avg.head()
```

```python
plt.figure(figsize=(10,7))
plt.hist(orders_25_avg.avg_days_order_since_25, bins=100)
plt.show()
```

```python
avg_days_since_prior25 = orders_25_avg.avg_days_order_since_25.mean()
avg_days_since_priorall = orders_prior.days_since_prior_order.mean()
roc=((avg_days_since_prior25 - avg_days_since_priorall)/avg_days_since_priorall)*100
print( "The rate od change is " +str(roc)+ '%' )
```
